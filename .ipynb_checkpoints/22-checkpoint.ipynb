# Imports

```python
import ast
import pandas as pd
```

# Read the file and convert to a list

```python
with open('NAMES.txt') as f:
    lines = f.readlines()

nhew = '[' + lines[0] + ']'
data = ast.literal_eval(nhew)
```

# Sort the list alphabetically

```python
data.sort()
```

# Build the abc dictionary

```python
abc = 'abcdefghijklmnopqrstuvwxyz'

d = {}

for i in range(1,len(abc)+1):
    d[abc[i-1].upper()] = i
    # print(d)
```

```python
index = 1

result_list = []
for name in data:
    score = 0
    for letter in name:
        score += d[letter]
        
    result_name = score * index
    result_list.append(result_name)
    
    # Done with the index * the name score
    index += 1



df = pd.DataFrame({
    'score': result_list,
    'name': data
})

df['score'].sum()
```
